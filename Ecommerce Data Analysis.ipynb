```python
#Pulling in the dataset for analysis
data = pd.read_excel('Online Retail.xlsx')
data
```

```python
#There was a problem with rows having NaN value in Description column
#there also may be an issue with int values
data = data.dropna(subset=['Description'])
```

```python
#Getting rid of rows where the value is not the right type
for val, row in data.iterrows():
    if type(row['Description']) == int or type(row['Description']) == float:
        data.drop(val, inplace=True)
        print(row['Description'])
        continue
    else:
#         data.at[val, 'desc word count'] = len(row['Description'].split())
        continue
```

```python
#to answer some of the questions, we need an item order total column
data['ItemOrderTot'] = 0
```

```python
#this creates the item order total for each row
for val, row in data.iterrows():
    temp = row.Quantity * row.UnitPrice
    data.at[val, 'ItemOrderTot'] = temp
```

```python
#created a series with customer ID of 100 best customers by order volume. 
best_customers = data.groupby("CustomerID").InvoiceNo.count().sort_values(ascending=False).head(100)
best_customers
```

```python
#this gives you best customers by revenue
best_customers_byRev = data.groupby("CustomerID").ItemOrderTot.sum().sort_values(ascending=False).head(10)
best_customers_byRev
```

```python
# these are the customers that are the in top 10 by revenue and order count
rev_invCount_customers = [x for x in best_customers.keys() if x in best_customers_byRev]
rev_invCount_customers
```

```python
#this gives you best customers by number of items purchased
best_customers_byItemCount = data.groupby("CustomerID").Quantity.sum().sort_values(ascending=False).head(10)
print(best_customers_byItemCount)

#is there overlap with best customers by revenue
overlap_itemCount_rev = [x for x in best_customers_byItemCount.keys() if x in best_customers_byRev]
print('Customers in top 10 of revenue and total items purchased: ', '\n', overlap_itemCount_rev)
```
